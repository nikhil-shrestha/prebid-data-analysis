```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# configuration
N = 300
D = 2
K = 3
```

```python
means = np.array([
  [0, 0],
  [0, 5],
  [5, 5],
])
```

```python
X = np.zeros((N, D))
X[:100, :] = np.random.randn(100, D) + means[0]
X[100:200, :] = np.random.randn(100, D) + means[1]
X[200:, :] = np.random.randn(100, D) + means[2]
```

```python
Y = np.zeros(N)

for n in range(N):
  closest_k = -1
  min_dist = float('inf') # initialize variable as infinity

  for k in range(K):
    d = (X[n] - means[k]).dot(X[n] - means[k])
    if d < min_dist:
      min_dist = d
      closest_k = k
  
  Y[n] = closest_k
```

```python
# visualize the data
plt.scatter(X[:, 0], X[:, 1], c=Y);
```
